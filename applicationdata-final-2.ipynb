### How to impute the missing data?
</li><li>Based on the above boxplots we can see that the **6 columns** with quantitative data have more than **13% missing values** and many outliers in the data,<br>
</li><li>Since, the outliers affect the mean values, hence we can't use mean for imputing the missing values. Instead, we could use **median** or **mode** to impute the missing values.
</li><li>In this case, we can use ***median*** as the outliers span over a huge distance over the boxplot outlier limits.

## Handling Negative Values in the Dataframe


### Inference for DAYS_BIRTH Column

</li><li>Here we can notice that the DAYS_BIRTH column defines the days past since a person was born, this data is not just confusing but can also lead to multiple errors during computations.

</li><li>Hence, we will be converting the **DAYS_BIRTH** column to **Age**, in order to define the age of any given user.

```python
# Checking the DAYS_BIRTH Column

application_data.DAYS_BIRTH
```

```python
# Calculating the Age of the customer from DAYS_BIRTH and storing it in the Age column

application_data["Age"]= abs((application_data['DAYS_BIRTH']/365 )).astype(int)
application_data["Age"]
```

### Inference for DAYS_EMPLOYED Column


</li><li>Here we can notice that the DAYS_EMPLOYED column defines the number of days an employee has been employed for.

</li><li>Hence, we will be converting the **DAYS_EMPLOYED** column to **Work_experience**, in order to define the experience of any given user.

```python
# Calculating the work experience of the customer from DAYS_EMPLOYED and storing it in the Work_Experience Column

application_data["Work_Experience"] = abs(round(application_data['DAYS_EMPLOYED'] / 365, 2))
application_data["Work_Experience"]
```
