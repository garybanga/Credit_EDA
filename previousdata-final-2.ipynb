### HOW DO WE IDENTIFY OUTLIERS HERE?
<ol><li> First of all we make use of boxplots to see if we can spot the outliers</li><li> Secondly, we have used the quantile method to identify the values of outliers. </li><li> In most cases anything above 99% quantile is an outliers and we remove them for the better analysis of our data

```python
# Checking for the outliers in the AMT_ANNUITY section

plt.figure(figsize=(15,5))

sns.boxplot(previous_data.AMT_ANNUITY)
plt.show()
```

```python
# Treating Outliers
previous_data.AMT_ANNUITY.quantile([0.5,0.7,0.8, 0.9,0.95,0.99])
```

```python
# Treating Outliers

previous_data.AMT_ANNUITY=previous_data.AMT_ANNUITY[previous_data.AMT_ANNUITY<70000]
```

```python
previous_data.AMT_ANNUITY.describe()
```

```python
# AMT_APPLICATION boxplot after removing outliers

plt.figure(figsize=(10,5))
sns.boxplot(previous_data.AMT_ANNUITY)
plt.show()
```

#### Inference
<ol><li>As we can see that the outliers have been treated but their are still some continous outliers making a line outside the quartile range. Its not required to removed them as its ueful for data analysis .</li><li>From the boxplot we can see that most of the data lies within 6000 to 20000</li><li> The mix value is 0 while the max value is approximately 70000
